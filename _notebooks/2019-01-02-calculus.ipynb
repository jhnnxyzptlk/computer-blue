## Overview

Calculus, sometimes known as infinitesimal calculus in its early days, is a mathematical subject that focuses on limits, continuity, derivatives, integrals, and infinite series. 

> Note: [History of Calculus](https://en.wikipedia.org/wiki/History_of_calculus)

We use two central ideas when dealing with calculus,  differential calculus and integral calculus. The former is concerned with instantaneous rates of change and curve slopes, whereas integral calculus is concerned with quantity accumulation and areas under or between curves. 

> Differential calculus can be thought of as a process of infinitely fine subtraction that is used to measure the differences between its parts. 

> Integral Calculus can be thought of as the process of infinitely fine additions, which integrates the parts back into the original whole.

## Limits and Continuity

> What does it mean to be continous

Our experience in our daily lives appears to be continuous. That is, there are no abrupt changes or discontinuities in what we perceive and experience. That's great news. Imagine you're out for a beautiful walk when all of a sudden a split second disappears from the cosmos. That would be disastrous and unexpected for our experience. 

> Note: No change in nature can ever take place “by a leap". 

Does continuity imply infinity? Well if we imagine a number line we assume there is an infinite number of numbers. That is there is a number in every posible position on the number line, its therefore smooth and continous. Yet within the infinite there seems to be discrete amount of numbers, such as the integers. What gives? 

> Note: I will write more on the `Principal of Continuity` in a future blog post. 

A rigorous definition of continuity of real functions is usually given in a first course in calculus in terms of the idea of a limit. In mathematics, many functions have the property that their graphs can be traced with a pencil without lifting the pencil from the page. Such functions are called continuous. 

Not all functions are continuous. Sometimes we cant figure out what a function is doing directly at a point. We can however see what it should be as we get closer to a limiting value.

## Derivatives

> what is a derivative? 

![](https://imgs.xkcd.com/comics/newton_and_leibniz.png)

The derivivative of a function measures the sensitivity to change of the function value (output value) with respect to a change in its argument (input value){% fn 1 %}.

What does this all mean? How can we vizualize this? 

We may approach this question from multiple perspectives.

### Geometric Interpretation of Differentiation

From a geometric point of view the derivative is the `slope` of the `tangent line` to the graph of $f(x)$. But what is a tangent line, exactly?

> It is the limit of the `secant line` (a line drawn between two points on the graph) as the distance
between the two points goes to zero.

![](https://ocw.mit.edu/courses/mathematics/18-01-single-variable-calculus-fall-2006/18-01f06.jpg)

You take the Point Q and you move it closer to P. In a more formal way but not rigorous, it is the limit of the slope of the secant line PQ as Q → P (where P fixed). The slope of PQ is represented by this graph: 

![](https://lh5.googleusercontent.com/zbxAFSCDEX8ssMLsrA-c3mG1TvylnagiMZobMlMhHv-bf977pnRRuak_eBDgdtteVJm8hYgOVzNoJFBQUDQRCo2qwa3ljGPMsI02GipnY7yCL4Yr5guXB5lu0Q)

As $Q$ approaches $P$ – i.e., as $x$ approaches $0$ from both sides – this quantity becomes the `instantaneous rate of change` at $P$, defined by:

$$\lim_{\Delta x \to 0} \frac{\Delta f}{\Delta x} = \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x} = f'(x)$$

>tip: This is called the difference quotient

### Physical Interpretation of Differentiation

One interpretation everyone is familar with is driving a car. You step on the pedal, the car accelerates. Your odometer tells you how fast you are going at any given moment in time. That seems kind of odd since the difference quotient implies the average of two points. Theres no trickery here, your odometer is calculating your speed with two points really really close together.

## Integrals

> Integration is the process of finding the antiderivative of a function.

Integration can be used to solve the area of an arbitrary shape, the length of a curve, and the volume of a solid, among others. 

> tip: The Area is a visualization technique for learning

![](https://image.shutterstock.com/image-vector/definite-integral-using-graph-on-260nw-1691608849.jpg)

$\int f(x) \,dx$

From a geometric point of view, the integral can be interpreted as the area under the curve of the mathematical function $f(x)$ plotted as a function of $x$. 

You may imagine drawing a huge number of blocks to approximate the area under a complex curve, with more blocks yielding a better solution. The integral provides a mathematical method for drawing an unlimited number of blocks and obtaining an exact analytical expression for the area.

![](https://image.shutterstock.com/image-vector/vector-graph-demonstration-integration-mathematical-600w-1883366197.jpg)

That's crucial for geometry, and even more so for the physical sciences, where the definitions of many physical phenomena, such as the area under a curve, can be expressed mathematically.

The area of a small block under a curve can be thought of as the width of the strip multiplied by (i.e., weighted by) the height of the strip. Many features of continuous substances are dependent on weighted sums, which must be infinite weighted sums to be correct - a challenge tailor-made for the integral.

> tip: `Riemann` and `Lebesgue` integrals are the most widely used definitions of the integral. 

This is a highly oversimplified overview of integrals. You can check out this video from 3b1b on [integrals](https://youtu.be/rfG8ce4nNh0). 

## The Fundamental Theorem of Calculus

The two branches are connected by the fundamental theorem of calculus, which shows how a definite integral is calculated by using its antiderivative (a function whose rate of change, or derivative, equals the function being integrated).
