## Proof

A proof is a method for ascertaning the truth.

How is truth established:
- Experiment and observation
- Sampling and counter examples
- Judge, jury
- Word of God

A mathematical proof is a verication of a `proposition` by a chain of `logical deductions` from a set of of `axioms`. 


## Proposions

A proposition is a statement that is either true or false

### Examples of simple propositions 

> 2 + 2 = 4

> $\forall n\ \in \mathbb{N}, n^2+n=41$ is a prime number

The first example, $2 + 2 = 4$,   . Reading the second example in English: For every nonnegative integer, $n$, the value of $n^2+n=41$ is prime. 
Lets break down the sentence structure and what these symbols mean before we continue. 

- $\forall$: for all, or 'for every possible choice of $n$

- $\in$: is a member of; belongs to; is in. 

- $\mathbb{N}$: The set of Natural numbers, {0, 1, 2, 3, ...}

- $n^2+n=41$ is a prime number: this part is called the `Predicate`, a proposition whose truth depends on the value of the variable. {% fn 1 %}
