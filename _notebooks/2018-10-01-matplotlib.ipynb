## Overview

The python ecosystem gives us the option of visualizing relationships between numbers with `matplotlib`, a plotting library which can produce quality figures. 

There are numerous plotting libraries available for python, such as `viola` and `seaborn`. 


```python
import matplotlib.pyplot as plt

def plot_ex(cxs, cys, dxs, dys):
    plt.xlabel('x', fontsize=20)
    plt.ylabel('f(x)', fontsize=20)
    plt.plot(cxs, cys, 'r-', label='quadratric')
    plt.plot(dxs, dys, 'b-', label='other function')
    plt.legend()
    plt.show()
    
cxs = [0.1*i for i in range(60)]
cys = [x**2 for x in cxs]
dxs = [i for i in range(7)]
dys = [x**1.8 - 0.5 for x in dxs]

plot_ex(cxs, cys, dxs, dys)
```
