# Overview

The method of calculating the value of the derivative of an explicitly undefined function with a given discrete set of points is known as numerical differentiation. 

One means of finding the derivative is to approximate the function locally by a polynomial and the to differentiate it. 

And equally effective tool is the taylor series expansion of a function about a point. 

To numerically distinguish a function, we first compute an interpolating polynomial and then compute the approximate derivative at the specified point.

> note: It is assumed that you already know about derivatives from an elementary calculus course. 

## The Derivative

Recall that the derivative $f'(x)$ of a function $f(x)$ at a point *x* is defined by the diffence quotient, where the limit of $\delta x$ tend to zero.

$$f'(x) = \lim_{\delta x \to 0} \frac{f(x + \delta x) - f(x)}{\delta x}$$

In other words, the difference quotient is an approximation of the derivative $f'(x)$, and this approximation gets better as $\delta x$ gets smaller.
