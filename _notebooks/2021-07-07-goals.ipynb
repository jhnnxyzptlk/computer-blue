   My adventure towards technology began roughly two years ago with Lambda School. I had no desire to return to a typical university for a bachelor's degree. I also lacked both time and funds. As a result, Lambda School sounded like a wise investment. I was mistaken. Now, I won't go into too much detail regarding Lambda School, but at the end of the program, I felt as though I still lacked *something*. Or perhaps the motivation for the type of computational skills I was seeking was misaligned. 

   So I spent a few months exploring as best I could on my computer screen (because of covid lockdowns). I discovered a plethora of exciting courses, I attended interesting online conferences, and listened to an unknown number of podcasts. One particular [Course](https://safari.ethz.ch/digitaltechnik/spring2021/doku.php) hosted by Professor [Onur Mutlu](https://people.inf.ethz.ch/omutlu/) caught my attention. The concept of learning about computer architecture from electrons to the problem you're attempting to solve is maybe the best expression of what I'm attempting to achieve: to do *great work* using the art of science and engineering for the betterment of humanity.
    
   Returning to the *something* that was lacking from Lambda School. Today, information is abundant, but wisdom is exceedingly rare. While Lambda School teaches the tools well, they lack the *soul*, or the *art* of the craft. They overlooked the nuances that enable Machine Learning to perform so well now. And that necessitates a great deal of mathematics, not less. We are facing a highly technical future in which CPU chips will be ubiquitous. In order to do outstanding work in tackling humanity's most pressing concerns, we will require specialist engineers who understand the entire stack, from electrons to the problem you're trying to solve. 

   For now, I'm mostly concentrating on completing Professor Onur Mutlu two excellent courses: [Digital Design and Computer Architecture](https://safari.ethz.ch/digitaltechnik/spring2021/doku.php) to start, and his more specialized [Computer Architecture](https://safari.ethz.ch/architecture_seminar/spring2021/doku.php) course, which covers the more cutting edge research in the field. You can check out more lecture videos [here](https://people.inf.ethz.ch/omutlu/lecture-videos.html). 
   
   As to which industry that I see with the biggest upside and in which I see myself working in, finance. Modern finance, or finance technology(finteck) is arguabily the most interesting industry today. I will be making a seperate post later to give you my full analysis on fintech. 

   Enjoy the rest of your summer. Covid-19 left me extremely drained emotionally and mentally. Thats why I decided on a sabbatical. I will return September 21, 2021, and finish the rest of the year strong.
